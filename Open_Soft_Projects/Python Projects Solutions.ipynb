```python
#Prime factorization

print("Enter the number whose prime factors you want to find!(greater than 0 obviously!)")
num = int(input())

listofprimes=[]
np = 0

for i in range(1,num+1):
    
     #print(f"i={i}")
    count=0
    
    for j in range(1,i+1):
         #print(f"j={j}")
        if i%j == 0:
            count+=1
    
    if count==2:
        #print(f"prime found = {i}")
        listofprimes.append(i)
        np += 1
         #print(f"np = {np}")

print("\nPrime factors found are:")
            
for i in range(0,np):
    if num%listofprimes[i] == 0:
        print(listofprimes[i])
        
        
```
